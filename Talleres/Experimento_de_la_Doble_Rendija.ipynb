# 1. Introducción

El experimento de la doble rendija es un pilar fundamental en la física cuántica, demostrando la naturaleza dual onda-partícula de la luz (y la materia). Cuando la luz pasa a través de dos pequeñas rendijas, no solo produce dos puntos brillantes, sino un patrón de interferencia de múltiples franjas claras y oscuras en una pantalla. Esto sucede porque las ondas de luz se difractan al pasar por cada rendija y luego se superponen, reforzándose (interferencia constructiva) o cancelándose (interferencia destructiva) mutuamente.

Este documento presenta la recreación de este histórico experimento, incluyendo su montaje físico, los resultados obtenidos y una simulación computacional para visualizar y comprender mejor los principios ondulatorios involucrados.

## 4.2. Código de Simulación

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parámetros del experimento ---
wavelength = 650e-9  # Longitud de onda del láser (650 nm para un láser rojo)
slit_separation = 0.05e-2  # Separación entre rendijas (0.05 cm = 0.0005 m)
slit_width = 40e-6   # Ancho de cada rendija (40 micrómetros = 0.00004 m, valor estimado)
distance_to_screen = 1.0  # Distancia de las rendijas a la pantalla (1 metro)

# --- Rango de la pantalla para la simulación ---
y_max = 0.05  # Altura máxima en la pantalla (0.05 metros = 5 cm a cada lado del centro)
y = np.linspace(-y_max, y_max, 1000)  # Puntos en la pantalla

# --- Cálculos angulares ---
theta = np.arctan(y / distance_to_screen)

# --- Términos de Difracción (single slit) e Interferencia (double slit) ---
# Término de difracción (beta en la fórmula usual, pero aquí lo llamamos alpha_val)
alpha_val = (np.pi * slit_width * np.sin(theta)) / wavelength
single_slit_diffraction = (np.sin(alpha_val) / alpha_val)**2
# Manejar la división por cero para alpha_val = 0 (centro)
single_slit_diffraction[alpha_val == 0] = 1

# Término de interferencia (gamma en la fórmula usual, pero aquí lo llamamos beta_val)
beta_val = (np.pi * slit_separation * np.sin(theta)) / wavelength
double_slit_interference = np.cos(beta_val)**2

# --- Intensidad total ---
intensity = single_slit_diffraction * double_slit_interference

# --- Normalizar la intensidad para una mejor visualización ---
intensity_normalized = intensity / np.max(intensity)

# --- Visualización ---
plt.figure(figsize=(10, 6))
plt.plot(y * 100, intensity_normalized, color='blue')
plt.title('Simulación del Patrón de Interferencia de Doble Rendija')
plt.xlabel('Posición en la pantalla (cm)')
plt.ylabel('Intensidad de la luz (Normalizada)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(-y_max * 100, y_max * 100) # Asegurar el límite del eje x en cm
plt.show()

# --- Explicación de los parámetros para la documentación ---
print("\n--- Parámetros de la simulación ---")
print(f"Longitud de onda del láser (λ): {wavelength * 1e9:.0f} nm")
print(f"Separación entre rendijas (d): {slit_separation * 1e3:.2f} mm")
print(f"Ancho de cada rendija (a): {slit_width * 1e6:.0f} µm")
print(f"Distancia a la pantalla (L): {distance_to_screen:.1f} m")
```

## 4.3. Explicación del Código

1. Importar Librerías: Se importan numpy para operaciones numéricas eficientes y matplotlib.pyplot para graficar los resultados.

2. Parámetros del Experimento:

* wavelength: La longitud de onda del láser, crucial para la escala del patrón. Un láser rojo común tiene ~650 nm.

* slit_separation: La distancia entre los centros de las dos rendijas, que es uno de los parámetros que usted especificó (0.05 cm).

* slit_width: El ancho de cada rendija individual. Aunque no lo especificó, es un parámetro importante para el efecto de difracción. Un valor típico es de decenas de micrómetros.

* distance_to_screen: La distancia desde las rendijas hasta donde se proyecta el patrón. Este valor impacta directamente en el tamaño aparente del patrón.

3. Rango de la Pantalla: Se define el rango de la pantalla (y) donde se calculará la intensidad de la luz, creando un array de puntos donde se evaluará la intensidad.

4. Cálculos Angulares: Se calcula el ángulo (θ) para cada punto y en la pantalla.

5. Términos de Difracción e Interferencia:

* Difracción de una sola rendija: Se calcula el factor de difracción, que describe la envoltura general del patrón. Este término hace que la intensidad disminuya a medida que nos alejamos del centro.

* Interferencia de doble rendija: Se calcula el factor de interferencia, que produce las franjas claras y oscuras dentro de la envoltura de difracción.

* Se incluye una línea para manejar el caso especial donde alpha_val es cero (en el centro del patrón) para evitar errores de división por cero.

6. Intensidad Total: Ambos términos se multiplican para obtener la intensidad total, que muestra cómo la difracción de las rendijas individuales modula el patrón de interferencia de doble rendija.

7. Normalización: La intensidad se normaliza dividiéndola por su valor máximo para que la gráfica sea más fácil de interpretar (la intensidad máxima será 1).

8. Visualización: matplotlib se utiliza para graficar la intensidad normalizada en función de la posición en la pantalla, mostrando el patrón de interferencia simulado.
