## Complex Internal Product for Column Vectors

# Properties of the Internal Product in Vector Spaces

The internal product (or inner product) is a key concept in the study of vector spaces, providing a way to define geometric concepts such as length, angle, and orthogonality within these spaces. The internal product for a vector space is defined as a function:

$$ \langle {-},{-} \rangle : \mathbb{V} \times \mathbb{V} \rightarrow \mathbb{C}  $$

that satisfies the following properties for any vectors $x$, $y$, $z$ in the vector space $V$ and any scalar $c$:

1. Nondegenerate
   $$ \langle {V},{V} \rangle \geq 0  $$
   $$ \langle {V},{V} \rangle = 0 \text{ si solo si }V = \boldsymbol{0} $$
3. Respects addition:
   $$ \langle {V_1 + V_2},{V_3} \rangle = \langle {V_1},{V_3} \rangle + \langle {V_2},{V_3} \rangle  $$
   $$ \langle {V_1},{V_2 + V_3} \rangle = \langle {V_1},{V_2} \rangle + \langle {V_1},{V_3} \rangle $$
5. Respects scalar multiplication
   $$ \langle {c \cdot V_1},{V_2} \rangle = c \times \langle {V_1},{V_2} \rangle $$
   $$ \langle {V_1},{c \cdot V_2} \rangle = \overline{c} \times \langle {V_1},{V_2} \rangle$$
7. Skew symmetric:
   $$ \langle {V_1},{V_2} \rangle = \overline{\langle {V_2},{V_1} \rangle} $$

The definition of the internal product varies across different vector spaces.


## Example 1: Internal Product in a Complex Vector Space of Column Vectors

In a complex vector space of column vectors $\mathbb{C}^n$, the internal product (often called the dot product in real vector spaces) is defined as follows for vectors $ \mathbf{a} = [a_1, a_2, \ldots, a_n]^T $ and $ \mathbf{b} = [b_1, b_2, \ldots, b_n]^T $:

$$ \langle V_1, V_2 \rangle = V_1^\dagger \star V_2 = \sum_{i=0}^{n-1} \overline{V_1[i]} \times V_2[i] $$

where $ \overline{V_1[i]} $ denotes the complex conjugate of $ V_1[i] $.

### Example:

Given $ \mathbf{a} = [1+i, 2-i]^T $ and $ \mathbf{b} = [2+i, 3+i]^T $, the internal product is:

$$ \langle \mathbf{a}, \mathbf{b} \rangle = (1-i)(2+i) + (2+i)(3+i) = (3-i) + (5 +5i) = 8 + 4i $$

### Example in Python


```python

import numpy as np

# Define two complex column vectors
vector_a = np.array([1+1j, 2-1j])
vector_b = np.array([2+1j, 3+1j])

# Compute the complex internal product
complex_internal_product_vector = np.vdot(vector_a, vector_b)

complex_internal_product_vector
```

## Example 2: Internal Product in a Complex Vector Space of Square Matrices

For square complex matrices, the internal product is defined for matrices $A$ and $B$ as:

$$ \langle A, B \rangle = \text{Tr}(\bar{A}^T B) $$

where $ \text{Tr} $ denotes the trace of a matrix (the sum of all diagonal elements) and $ \bar{A}^T $ is the conjugate transpose of $ A $.

### Example:

Given

$$ A = \begin{bmatrix} 1+i & 2 \\ 3 & 4-i \end{bmatrix}, \quad B = \begin{bmatrix} 5 & 6+i \\ 7-i & 8 \end{bmatrix} $$

the internal product is:

$$ \langle A, B \rangle = \text{Tr} \left( \begin{bmatrix} 1-i & 3 \\ 2 & 4+i \end{bmatrix} \begin{bmatrix} 5 & 6+i \\ 7-i & 8 \end{bmatrix} \right) = \text{Tr} \left( \begin{bmatrix} 26-8j & 31-5j \\ 39 + 3j & 44 + 10j \end{bmatrix} \right) = 70 +2j $$

Note: The exact value of the trace depends on the computation of the product and the sum of the diagonal elements.

### The example in Python:


```python
import numpy as np

# Define two square complex matrices
matrix_A = np.array([[1+1j, 2], [3, 4-1j]])
matrix_B = np.array([[5, 6+1j], [7-1j, 8]])

# Compute the complex internal product for matrices
print(np.dot(np.conjugate(matrix_A).T, matrix_B))
complex_internal_product_matrix = np.trace(np.dot(np.conjugate(matrix_A).T, matrix_B))

complex_internal_product_matrix
```

# Eigenvalues and Eigenvectors

Below are examples that illustrate how to compute eigenvalues, eigenvectors, and how to plot matrix transformations in a Jupyter Notebook using Python. These examples utilize the numpy and matplotlib libraries, so make sure these are installed in your environment.

```python
import numpy as np

# Define a square matrix
A = np.array([[4, 2], [1, 3]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

This code snippet computes the eigenvalues and eigenvectors of the matrix A and prints them. The np.linalg.eig function returns a tuple consisting of an array of eigenvalues and a 2D array of the corresponding eigenvectors.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a transformation matrix
T = np.array([[2, 1], [1, 3]])

# Generate a set of vectors (unit circle)
theta = np.linspace(0, 2*np.pi, 100)
vectors = np.array([np.cos(theta), np.sin(theta)])

# Apply the transformation
transformed_vectors = T.dot(vectors)

# Plotting
plt.figure(figsize=(6,6))
plt.plot(vectors[0], vectors[1], 'b', label='Original Vectors')  # Original vectors
plt.plot(transformed_vectors[0], transformed_vectors[1], 'r', label='Transformed Vectors')  # Transformed vectors
plt.axis('equal')
plt.legend()
plt.title('Matrix Transformation')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

# Exercises 1-2

```python
import numpy as np

print("=== EJERCICIOS DE PRODUCTO INTERNO COMPLEJO ===")

# Ejercicio 1: Producto interno para vectores columna
print("\n--- Ejercicio 1: Producto Interno de Vectores ---")
a = np.array([[1 + 2j], [3 - 4j]])
b = np.array([[2 - 1j], [-1 + 3j]])

# Clculo manual del producto interno
# a,b = ab = sum(conj(a[i]) * b[i])
producto_interno_manual = np.conj(a[0,0]) * b[0,0] + np.conj(a[1,0]) * b[1,0]

# Usando numpy
producto_interno_numpy = np.vdot(a.flatten(), b.flatten())

print("Vector a:")
print(a)
print("Vector b:")
print(b)
print(f"Producto interno a,b (manual): {producto_interno_manual}")
print(f"Producto interno a,b (numpy): {producto_interno_numpy}")

# Verificacin paso a paso
print("\n--- Clculo Paso a Paso ---")
print(f"conj(a[0]) = conj({a[0,0]}) = {np.conj(a[0,0])}")
print(f"conj(a[1]) = conj({a[1,0]}) = {np.conj(a[1,0])}")
print(f"conj(a[0])  b[0] = {np.conj(a[0,0])}  {b[0,0]} = {np.conj(a[0,0]) * b[0,0]}")
print(f"conj(a[1])  b[1] = {np.conj(a[1,0])}  {b[1,0]} = {np.conj(a[1,0]) * b[1,0]}")

# Ejercicio 2: Producto interno para matrices
print("\n--- Ejercicio 2: Producto Interno de Matrices ---")
A = np.array([[1+1j, 2-2j],
              [3+3j, 4-1j]])
B = np.array([[1-2j, 3+1j],
              [4-4j, 2+2j]])

# Producto interno para matrices: A,B = Tr(AB)
A_adjoint = A.T.conj()
producto_AB = np.dot(A_adjoint, B)
producto_interno_matrices = np.trace(producto_AB)

print("Matriz A:")
print(A)
print("Matriz B:")
print(B)
print("A (adjunta de A):")
print(A_adjoint)
print("A  B:")
print(producto_AB)
print(f"Producto interno A,B = Tr(AB) = {producto_interno_matrices}")
```

```python
import matplotlib.pyplot as plt

print("\n--- Ejercicio 4: Visualizacin de Transformaciones ---")

# Matriz de transformacin
T = np.array([[2, 1],
              [1, 3]])

# Generar crculo unitario
theta = np.linspace(0, 2*np.pi, 100)
unit_circle = np.array([np.cos(theta), np.sin(theta)])

# Aplicar transformacin
transformed_circle = np.dot(T, unit_circle)

# Calcular eigenvalores y eigenvectores
eigenvalues_T, eigenvectors_T = np.linalg.eig(T)

print("Matriz de transformacin T:")
print(T)
print(f"\nEigenvalores: {eigenvalues_T}")
print("Eigenvectores:")
print(eigenvectors_T)

# Crear figura con 4 subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Subplot 1: Crculo original vs transformado
ax1 = axes[0, 0]
ax1.plot(unit_circle[0], unit_circle[1], 'b-', label='Crculo Original', linewidth=2)
ax1.plot(transformed_circle[0], transformed_circle[1], 'r-', label='Crculo Transformado', linewidth=2)
ax1.axis('equal')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_title('1. Transformacin del Crculo Unitario')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_xlim(-2, 6)
ax1.set_ylim(-2, 6)

# Subplot 2: Eigenvectores originales
ax2 = axes[0, 1]
ax2.plot(unit_circle[0], unit_circle[1], 'b-', alpha=0.3, label='Crculo Original')

# Dibujar eigenvectores originales con diferentes colores
colors = ['red', 'green']
for i, (, v) in enumerate(zip(eigenvalues_T, eigenvectors_T.T)):
    # Normalizar para mejor visualizacin
    v_norm = v / np.linalg.norm(v) * 1.5
    ax2.arrow(0, 0, v_norm[0], v_norm[1],
              head_width=0.1, head_length=0.1,
              fc=colors[i], ec=colors[i], linewidth=3,
              label=f'v{i+1}: ={:.2f}')
    # Aadir lnea extendida
    ax2.plot([-2*v_norm[0], 2*v_norm[0]], [-2*v_norm[1], 2*v_norm[1]],
             '--', color=colors[i], alpha=0.5)

ax2.axis('equal')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_title('2. Eigenvectores Originales')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(-2, 2)
ax2.set_ylim(-2, 2)

# Subplot 3: Eigenvectores transformados
ax3 = axes[1, 0]
ax3.plot(transformed_circle[0], transformed_circle[1], 'r-', alpha=0.3, label='Crculo Transformado')

# Dibujar eigenvectores transformados
for i, (, v) in enumerate(zip(eigenvalues_T, eigenvectors_T.T)):
    transformed_v = np.dot(T, v)
    # Normalizar el vector original para comparacin
    v_norm = v / np.linalg.norm(v) * 1.5
    transformed_v_display = np.dot(T, v_norm)

    ax3.arrow(0, 0, transformed_v_display[0], transformed_v_display[1],
              head_width=0.2, head_length=0.2,
              fc=colors[i], ec=colors[i], linewidth=3,
              label=f'Tv{i+1} = {:.2f}v{i+1}')

ax3.axis('equal')
ax3.grid(True, alpha=0.3)
ax3.legend()
ax3.set_title('3. Eigenvectores Transformados')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_xlim(-2, 6)
ax3.set_ylim(-2, 6)

# Subplot 4: Comparacin lado a lado
ax4 = axes[1, 1]
# Eigenvectores originales
for i, (, v) in enumerate(zip(eigenvalues_T, eigenvectors_T.T)):
    v_norm = v / np.linalg.norm(v) * 1.5
    transformed_v_display = np.dot(T, v_norm)

    # Vector original
    ax4.arrow(0, 0, v_norm[0], v_norm[1],
              head_width=0.1, head_length=0.1,
              fc=colors[i], ec=colors[i], linewidth=2, alpha=0.5,
              linestyle='--', label=f'Original v{i+1}')

    # Vector transformado
    ax4.arrow(0, 0, transformed_v_display[0], transformed_v_display[1],
              head_width=0.15, head_length=0.15,
              fc=colors[i], ec=colors[i], linewidth=3,
              label=f'Transformado v{i+1}')

ax4.axis('equal')
ax4.grid(True, alpha=0.3)
ax4.legend()
ax4.set_title('4. Comparacin: Original vs Transformado')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_xlim(-2, 6)
ax4.set_ylim(-2, 6)

plt.tight_layout()
plt.show()

# Verificacin numrica detallada
print("\n" + "="*60)
print("VERIFICACIN NUMRICA DETALLADA")
print("="*60)

for i, (, v) in enumerate(zip(eigenvalues_T, eigenvectors_T.T)):
    print(f"\nEIGENVECTOR {i+1}:")
    print(f"  Eigenvalor {i+1} = {:.6f}")
    print(f"  Eigenvector v{i+1} = [{v[0]:.6f}, {v[1]:.6f}]")

    # Calcular T  v
    Tv = np.dot(T, v)
    print(f"  T  v{i+1} = [{Tv[0]:.6f}, {Tv[1]:.6f}]")

    # Calcular   v
    v =  * v
    print(f"  {i+1}  v{i+1} = [{v[0]:.6f}, {v[1]:.6f}]")

    # Verificar igualdad
    es_igual = np.allclose(Tv, v, rtol=1e-10)
    print(f"  Tv = v? {es_igual}")

    if es_igual:
        print(f"   El vector v{i+1} es escalado por un factor de {:.3f}")
    else:
        print(f"   Error en el clculo")

print("\n" + "="*60)
print("INTERPRETACIN GEOMTRICA")
print("="*60)
print("\nLa matriz T transforma el crculo unitario en una elipse porque:")
print("1. Los eigenvalores son diferentes (  ), causando escalamiento desigual")
print("2. Los eigenvectores no son ortogonales, causando rotacin")

print(f"\nDireccin de mayor estiramiento: eigenvector con  = {max(eigenvalues_T):.3f}")
print(f"Direccin de menor estiramiento: eigenvector con  = {min(eigenvalues_T):.3f}")

print("\nLos eigenvectores indican las direcciones principales donde la transformacin")
print("solo escala (no rota) los vectores, multiplicndolos por sus eigenvalores correspondientes.")
```
